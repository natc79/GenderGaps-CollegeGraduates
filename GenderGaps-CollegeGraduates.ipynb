# Gender Gaps in College Graduates (1970-2012)

This data comes from the Department of Education Statistics which contains the percentage of bachelor's degrees granted to women by year. The data set contains 17 categories of degrees, with each column as indicating a seperate degree category.  This dataset came from Randal Olson of the University of Pennsylvania.  The below analysis provides a visual representation of the different degree categories over time.

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']

fig = plt.figure(figsize=(18, 3))

def graphcharacteristics():
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[sp])
    ax.tick_params(bottom="off", top="off", left="off", right="off")

for sp in range(0,6):
    ax = fig.add_subplot(1,6,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    graphcharacteristics()
    if sp == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif sp == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
plt.show()
```

```python
stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering', 'Computer Science']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']

fig = plt.figure(figsize=(24, 12))

def graphcharacteristics(sp):
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    #Enable y-axis labels only at 0 and 100
    ax.set_yticks([0,100])
    #Provide a horizontal line across subplot at mid-point
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    #diables all tick marks except for the bottom most line charts
    if sp < 5:
        ax.tick_params(bottom="off", top="off", left="off", right="off", labelbottom="off")
    elif sp == 5:
        ax.tick_params(labelbottom="on")
        
for sp in range(0,6):
    #adding in plots for stem_cats
    ax = fig.add_subplot(6,3,sp*3+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    graphcharacteristics(sp)
    ax.set_title(stem_cats[sp])
    if sp == 0 or sp == 5:
        menp = 105-women_degrees[stem_cats[sp]][women_degrees['Year']==2005]
        womenp = 5 + women_degrees[stem_cats[sp]][women_degrees['Year']==2002]
        ax.text(2005, menp, 'Men')
        ax.text(2002, womenp, 'Women')
    #adding in plots for lib_arts_cats
    if sp < 5:
        ax = fig.add_subplot(6,3,sp*3+2)
        ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
        ax.plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[sp]], c=cb_orange, label='Men', linewidth=3)
        graphcharacteristics(sp)
        ax.set_title(lib_arts_cats[sp])
    if sp == 0:
        menp = 105-women_degrees[lib_arts_cats[sp]][women_degrees['Year']==2005]
        womenp = 5 + women_degrees[lib_arts_cats[sp]][women_degrees['Year']==2002]
        ax.text(2005, menp, 'Men')
        ax.text(2002, womenp, 'Women')
    #adding in plots for other cats
    ax = fig.add_subplot(6,3,sp*3+3)
    ax.plot(women_degrees['Year'], women_degrees[other_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[other_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    graphcharacteristics(sp)
    ax.set_title(other_cats[sp])
    if sp == 0 or sp == 5:
        menp = 105-women_degrees[other_cats[sp]][women_degrees['Year']==2005]
        womenp = 5 + women_degrees[other_cats[sp]][women_degrees['Year']==2002]
        ax.text(2005, menp, 'Men')
        ax.text(2002, womenp, 'Women')
plt.savefig("C:\JobApplications\PythonCoding\gender_degrees.png")
```

### Conclusions

We see over time there has been a narrowing in the gender gap for various college majors.  Nevertheless many majors still have strong gender differences especially is true for some of the STEM categories such as engineering.  Moreover, psychology majors have become increasingly dominated by women.
